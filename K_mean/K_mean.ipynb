```
def compute_center(X):
  # Computing center of X
  return np.mean(X,axis=0)
```

```
def distance(a,b, norm=2):
  """
  Compute the distance between a and b with norm
  
  Parameters:
    a,b - np.array : two points that need to be compute the distance between them
    norm - int : the norm - default = 2
    
  Returns:
    the distance
  """
  tmp = a - b
  if norm == 1:
    return sum(np.abs(tmp))
  else:
    return np.power(sum(np.power(tmp,norm)),1./norm)
```
